```python
import cv2
import matplotlib.pyplot as plt
```

```python
im1 = cv2.imread('im1.png')
im1 = cv2.cvtColor(im1, cv2.COLOR_BGR2RGB)

im2 = cv2.imread('im2.png')
im2 = cv2.cvtColor(im2, cv2.COLOR_BGR2RGB)

im3 = cv2.imread('im3.png')
im3 = cv2.cvtColor(im3, cv2.COLOR_BGR2RGB)

im4 = cv2.imread('im4.png')
im4 = cv2.cvtColor(im4, cv2.COLOR_BGR2RGB)
```

```python
draw1 = im1.copy()
cv2.circle(draw1, (1000, 890), 5, (0, 255, 0), 2)
plt.imshow(draw1)

# 1000, 890
```

```python
draw2 = im2.copy()
cv2.circle(draw2, (950, 450), 5, (0, 255, 0), 2)
plt.imshow(draw2)

# 950, 450
```

```python
draw3 = im3.copy()
cv2.circle(draw3, (950, 770), 5, (0, 255, 0), 2)
plt.imshow(draw3)
```

```python
draw4 = im4.copy()
cv2.circle(draw4, (1000, 750), 5, (0, 255, 0), 2)
plt.imshow(draw4)

# 1000, 750
```
