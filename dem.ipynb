```python
import cv2
import matplotlib.pyplot as plt
from reader import Reader
```

```python
img = cv2.imread('grid_image\img_battle.png')
_, (x, y), width = Reader.read(img)
print(x, y, width)

start_x, start_y = x + 20, y + 20
width = int(width)
draw = img.copy()

for row in range(16):
    for col in range(16):
        x = start_x + col * width
        y = start_y + row * width

        cv2.circle(draw, (x, y), 5, (0, 255, 0), 3)

plt.imshow(draw)
```
